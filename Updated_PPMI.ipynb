```python
import numpy as np
import pandas as pd

def ppmi(cooccurrence_df):
    #converts df to a form numpy can work with
    co_occurrence_matrix = cooccurrence_df.values

    #relevant sums
    total_occurrences = np.sum(co_occurrence_matrix)
    row_sum = co_occurrence_matrix.sum(axis=1)
    col_sum = co_occurrence_matrix.sum(axis=0)

    # Calculate the PPMI matrix
    pmi_matrix = np.log2((total_occurrences * co_occurrence_matrix) / (row_sum.reshape((-1, 1)) * col_sum))

    #replaces the ensuing nans(from zero division) with zeros
    pmi_matrix = np.nan_to_num(pmi_matrix, nan=0.0)

    #converts the pmi matrix into a ppmi matrix
    pmi_matrix[pmi_matrix < 0] = 0

    return pmi_matrix
```

```python
ppmi_matrix = ppmi(co_matrix_df)

```

```python
ppmi = pd.DataFrame(ppmi_matrix)
ppmi.to_csv('/content/drive/My Drive/IMT575Final/' + 'ppmi.csv') # maxwell
#ppmi.to_csv("/content/drive/My Drive/IMT575_HW1/ppmi.csv") #Ken's path
```
